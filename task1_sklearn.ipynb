```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

## read the dataset

```python
X_set = pd.read_csv('X_train.csv') #remove the headline automatically
Y_set = pd.read_csv('Y_train.csv')
print(X_set.shape, Y_set.shape)
X_set.head(3) #to check whether data read correctlly, return the first 3 rows of datasets
X_set.isnull().sum(0)
```

```python
# print(X_set.describe()) #show some statistics of dataset
```

```python
# separate the other attributes from the predicting attribute
X_set = X_set.drop(['id'], axis=1) #remove the id column
Y_set = Y_set['y'] #remove the id column

X_set.shape
X_set.head(3)
```

## Dataset processing

```python
# do Normalization
X_set = (X_set-X_set.min())/(X_set.max()-X_set.min())
# another normalization method
# X_set = (X_set-X_set.mean())/X_set.std()   
X_set.head(20)
```

```python
#use each col's mean to fill the NAN default value
# X_set = X_set.fillna(X_set.mean())
X_set = X_set.fillna(0)
X_set.head(7)
```

```python
print("X_set.shape:", X_set.shape)
print("Y_set.shape:", Y_set.shape)
```

```python
X_test = pd.read_csv('X_test.csv')
index = X_test['id']
X_test = X_test.drop(['id'], axis=1) #remove the id column
X_test = (X_test-X_test.min())/(X_test.max()-X_test.min())
X_test = X_test.fillna(0)
X_test.head()
```

## Regression Method

```python
LR = LinearRegression()


X_set.isnull().sum(0)
#X_set[X_set.isnull().values == True]
#X_set.shape
#print(X_set)
#print("\n\n",Y_set)
LR.fit(X_set,Y_set)
y_prediction = LR.predict(X_test)
Y_prediction = LR.predict(X_set)
#print(y_prediction)
prediction_results = pd.DataFrame(data=y_prediction,columns=['y'])
# Using DataFrame.insert() to add a column
index = [i for i in range(len(prediction_results))]
prediction_results.insert(0,"id",index)
  
# Observe the result
prediction_results
prediction_results.to_csv('result_0.csv',index = False)

prediction_results = pd.DataFrame(data=Y_prediction,columns=['Y'])
# Using DataFrame.insert() to add a column
index = [i for i in range(len(prediction_results))]
prediction_results.insert(0,"id",index)
  
# Observe the result
prediction_results
prediction_results.to_csv('result_1.csv',index = False)
```
